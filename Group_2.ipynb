### Replicate the graph

```python
la = ["2011-12","2012-13","2013-14","2014-15","2015-16","2016-17","2017-18","2018-19","2019-20"]

df = pd.DataFrame({"A":[84.6,69.6,84.3,77.8,60.9,128,164.8,116.9,141.7],"B":[19.3,15.7,18.5,16.3,13.5,22.3,26.6,19.5,24]})
index = [-1,0,1,2,3,4,5,6,7]

fig, ax1 = plt.subplots(figsize = (8,5))
ax1.set_xlim(0,7.5)
a = plt.bar(index, df.A, color = '#7092b7', width = 0.75,label = "Amount of land premium paid by developers")
ax1.set_ylim(0,200)
plt.yticks(ticks = [0,50,100,150,200], fontweight = BOLD)
ax2 = ax1.twinx()

ax2.plot(index, df.B,mec = "#c23428", mfc = "white", color = "#c23428",marker="o",ms = 16.5,linewidth=5, label = " % of total HK goverment revenue")
ax2.set_ylim(0,30)
plt.xticks(ticks = [-1.5,-0.5,0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5], color = "red",fontweight = BOLD)

ax1.tick_params(axis=u'both', which=u'both',length=5,labelbottom = False)
ax1.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)
plt.text(-1.4, 29.5, "Amount of land premium paid by developers", color = "#2a4968", size = 14, fontweight = BOLD)
ax2.legend(loc = 'upper left', bbox_to_anchor=(0,0.9),framealpha = 0)
ax1.set_title("HK$bn",x=-0.04,y=1.02, fontweight = BOLD)
ax2.set_title("%",x=1.02,y=1, color = "#c23428", fontweight = BOLD)


for i in range(7):
    ax2.get_yticklabels()[i].set_color("#c23428")
    ax2.get_yticklabels()[i].set_weight(BOLD)
for i in range(9):
    plt.text(i-1.39, -1.1, la[i],fontweight = BOLD, size = 9)
for i in range(9):
    if i > 4:
        if i == 5:
            plt.text(i-1.2, 2, str(int(df.A[i])), color = "white",fontweight = BOLD)
        else:
            plt.text(i-1.32, 2, str(df.A[i]), color = "white",fontweight = BOLD)
    else:
        plt.text(i-1.25, 2, df.A[i], color = "white", fontweight = BOLD)
db = [i+2 for i in df.B]
for i in range(9):
    if i == 4:
        plt.text(i-1.2, db[i]-4, str(df.B[i])+"%", color = "#c23428", fontweight = BOLD)
    elif i == 5:
        plt.text(i-1.4, db[i]-0.5, str(df.B[i])+"%", color = "#c23428", fontweight = BOLD)
    elif i == 6:
        plt.text(i-1.2, db[i]-1, str(df.B[i])+"%", color = "#c23428", fontweight = BOLD)
    elif i == 7:
        plt.text(i-1.2, db[i]+1, str(df.B[i])+"%", color = "#c23428", fontweight = BOLD)
    elif i == 8:
        plt.text(i-1.2, db[i]-1, str(int(df.B[i]))+"%", color = "#c23428", fontweight = BOLD)
    else:
        plt.text(i-1.2, db[i]-1, str(df.B[i])+"%", color = "#c23428", fontweight = BOLD)
%config InlineBackend.figure_format = 'svg'
plt.show()
```

```python
la = ["2011-12","2012-13","2013-14","2014-15","2015-16","2016-17","2017-18","2018-19","2019-20"]

df = pd.DataFrame({"A":[84.6,69.6,84.3,77.8,60.9,128,164.8,116.9,141.7],"B":[19.3,15.7,18.5,16.3,13.5,22.3,26.6,19.5,24]})
index = [-1,0,1,2,3,4,5,6,7]

fig, ax1 = plt.subplots(figsize = (10,5))

ax1.set_xlim(0, 7.5)

a = plt.bar(index, df.A, color = '#7092b7', width = 0.6, label=None)

ax1.set_ylim(0,200)
plt.yticks(ticks = [0,50,100,150,200])

plt.xticks(ticks = [-1.5,-0.5,0.5,1.5,2.5,3.5,4.5,5.5,6.5], labels = la, color = "red")

ax1.tick_params(axis=u'both', which=u'both',length=5,labelbottom = False)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
plt.yticks([])
pos=[87.6,72.6,87.3,80.8,63.9,131,167.8,119.9,144.7]
ax1.set_title("Amount of land premium paid by developers(HK$bn)",fontsize=15)
for i in range(9):
    plt.text(i-1.35, -8, la[i],fontweight = BOLD)
for i in range(9):
    if i > 4:
        plt.text(i-1.25, pos[i], df.A[i], color = "black")
    else:
        plt.text(i-1.2, pos[i], df.A[i], color = "black")

plt.show()

print("-"*90)
fig, ax2 = plt.subplots(figsize = (10,5))
ax2.plot(index, df.B,mec = "#7092b7", mfc = "white", color = "#7092b7",marker="o",ms = 16.5,linewidth=5)
ax2.set_xlim(0, 7.5)
ax2.set_ylim(5, 35)
plt.xticks(ticks = [-1.5,-0.5,0.5,1.5,2.5,3.5,4.5,5.5,6.5], labels = "")
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
plt.yticks([])
ax2.set_title(" % of total HK goverment revenue",fontsize=15)

for i in range(9):
    plt.text(i-1.35, 3.8, la[i],fontweight = BOLD)

db = [i+2 for i in df.B]
for i in range(9):
    if i == 4:
        plt.text(i-1.2, db[i]-4, str(df.B[i])+"%", fontweight = BOLD)
    else:
        plt.text(i-1.2, db[i], str(df.B[i])+"%", fontweight = BOLD)
plt.show()
```
